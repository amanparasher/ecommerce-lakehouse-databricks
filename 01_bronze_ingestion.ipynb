```python

# BRONZE LAYER — Raw Ingestion from Unity Catalog Volume
# Olist E-Commerce Dataset


# base path
RAW_PATH = "/Volumes/workspace/default/raw_data"

# bronze delta output path
BRONZE_PATH = "/Volumes/workspace/default/raw_data/bronze"

# Reading all 9 raw CSVs into DataFrames
orders_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_orders_dataset.csv")

customers_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_customers_dataset.csv")

order_items_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_order_items_dataset.csv")

payments_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_order_payments_dataset.csv")

reviews_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_order_reviews_dataset.csv")

products_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_products_dataset.csv")

sellers_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_sellers_dataset.csv")

geolocation_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/olist_geolocation_dataset.csv")

category_translation_raw = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{RAW_PATH}/product_category_name_translation.csv")

print("✅ All 9 CSVs loaded successfully")
```

```python
from pyspark.sql.functions import current_timestamp

# Adding ingestion timestamp to every table
def add_ingestion_metadata(df, source_name):
    return df.withColumn("_ingested_at", current_timestamp()) \
             .withColumn("_source_file", lit(source_name))

from pyspark.sql.functions import lit

orders_bronze      = add_ingestion_metadata(orders_raw, "olist_orders")
customers_bronze   = add_ingestion_metadata(customers_raw, "olist_customers")
order_items_bronze = add_ingestion_metadata(order_items_raw, "olist_order_items")
payments_bronze    = add_ingestion_metadata(payments_raw, "olist_payments")
reviews_bronze     = add_ingestion_metadata(reviews_raw, "olist_reviews")
products_bronze    = add_ingestion_metadata(products_raw, "olist_products")
sellers_bronze     = add_ingestion_metadata(sellers_raw, "olist_sellers")
geolocation_bronze = add_ingestion_metadata(geolocation_raw, "olist_geolocation")
category_bronze    = add_ingestion_metadata(category_translation_raw, "olist_category_translation")

print("✅ Metadata columns added to all tables")
```

```python
# changing all tables as Delta format to Bronze path
orders_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/orders")

customers_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/customers")

order_items_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/order_items")

payments_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/payments")

reviews_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/reviews")

products_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/products")

sellers_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/sellers")

geolocation_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/geolocation")

category_bronze.write.format("delta").mode("overwrite") \
    .save(f"{BRONZE_PATH}/category_translation")

print("✅ All 9 Bronze Delta tables written successfully")
print(f"📁 Location: {BRONZE_PATH}")
```
